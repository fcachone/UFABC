```
import numpy as np
import matplotlib.pyplot as plt
 
```

como fazer anotação na frente da linha de comando

1) $y=x^2$  


```
x= np.linspace(-150,200,150) #(intervalo, intervalo, pts grafico)
y= np.power(x,2) #comando e^x

plt.figure(dpi=75, facecolor = 'r') #dpi(tamanho do bloco) - cor
plt.plot(x,y,'b-',label= '$y=x^2$',linewidth= 2.33) #label (legenda)
plt.xticks(rotation=77) #rotação da coordenada x
plt.legend() #adcionar legenda no plot
```

2) $y=sin(x)$


```
x= np.linspace(-2*np.pi, 2*np.pi,75) #pi = np.pi
y= np.sin(x)

plt.figure(dpi=80, facecolor = 'c', edgecolor='w')
plt.plot(x,y, 'm-.' ,label='$y=sin(x)$',linewidth= 1.5)
plt.legend()
```

3) $y=x^2 -5x +6$


```
x= np.linspace(-75,25,700)
y= np.power(x,2) - np.multiply(x,5) + 6

plt.figure(dpi= 130, edgecolor= 'r')
plt.plot(x,y,'g-',label='$y=x^2 -5x +6$',linewidth= 0.4)
plt.legend()
plt.grid(color='k', linestyle= '-.', linewidth= 0.2)
```

4) $f_1(x)=\frac{x}{1+x^2}$




```
x= np.linspace(0,100,1000)
y= x/(1 + np.power(x,2))

plt.figure()
plt.plot(x,y,'c-',label='$f_1(x)$',linewidth= 1.0)
plt.legend(loc='best')
```

onde $f_1(x)= \frac{x}{1+x^2}$ representa a funcao fualana que estou desenhando...
